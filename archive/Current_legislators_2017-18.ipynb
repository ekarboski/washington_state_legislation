```python
def create_current_legislators_df(merged_final_df, legislator_df, biennium, date):
    '''
    Input
    ------
    merged_final_df: pandas dataframe loaded from wa_leg_staging database, merged_final table
    legislator_df: pandas dataframe loaded from wa_leg_staging database, legislator table
    biennium: str, example: '2017-18'
    date: str, example: '2018-03-01T00:00:00.000000000'
    
    
    Output
    ------
    current_legislator_df: pandas dataframe
    '''
    m_current = merged_final_df[merged_final_df['biennium'] == biennium]
    m_current = m_current[m_current['vote_date'] > date]
    current_legs = m_current['voter_id'].unique()
    
    current_dfs = []
    for leg in current_legs:
        leg_info = legislator_df[legislator_df['id'] == leg]
        current_dfs.append(leg_info)
    
    current_legislator_df = reduce((lambda x, y: x.append(y)), current_dfs)
    return current_legislator_df
```

```python
cur_leg_df = create_current_legislators_df(merged_final_df, legislator, 
                                           '2017-18','2018-03-01T00:00:00.000000000')
```
